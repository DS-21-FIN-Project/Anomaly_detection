```python
import math
import cv2
# 바운딩 박스 중점 구하는 함수
def calculate_center(box):
    x_center = (box[0] + box[2]) / 2
    y_center = (box[1] + box[3]) / 2
    return (x_center, y_center)


def calculate_font_center(box):
    x_center = (box[0] + box[2]) / 2 - 20
    y_max = min(box[1], box[3]) + 40
    return (x_center, y_max)


# 중점에서의 거리를 계산
def distance(c1, c2):
    return math.sqrt((c1[0] - c2[0])**2 + (c1[1] - c2[1])**2)


# 리스트 내의 좌표들중 거리가 가장 가까운 두 점의 인덱스 값을 찾는 함수
def find_closest_pair_indices(boxes):
    centers = [calculate_center(box) for box in boxes]
    min_distance = float('inf')
    closest_pair_indices = (-1, -1)
    
    # 각 중심점 쌍 사이의 거리를 계산
    for i in range(len(centers)):
        for j in range(i + 1, len(centers)):
            dist = distance(centers[i], centers[j])
            if dist < min_distance:
                min_distance = dist
                closest_pair_indices = (i, j)
    
    return closest_pair_indices

# 리스트 내의 좌표들중 거리가 가장 가까운 두 점의 좌표를 찾는 함수
def find_closest_pairs(boxes):
    closest_indices = find_closest_pair_indices(boxes)
    return (boxes[closest_indices[0]], boxes[closest_indices[1]])

# 두 좌표를 사용해서 만들 수 있는 가장 큰 좌표를 출력하는 함수
def create_largest_box(box1, box2):
    min_x = min(box1[0], box2[0])
    min_y = min(box1[1], box2[1])
    max_x = max(box1[2], box2[2])
    max_y = max(box1[3], box2[3])
    
    return [int(min_x), int(min_y), int(max_x), int(max_y)]


def display_action(frame, action_name, largest_box, box_color):
    cv2.rectangle(frame, (int(largest_box[0]), int(largest_box[1])), 
                  (int(largest_box[2]), int(largest_box[3])), box_color, 2)
    cv2.putText(frame, action_name, (int(calculate_font_center(largest_box)[0]), 
                                     int(calculate_font_center(largest_box)[1])), 
                cv2.FONT_HERSHEY_SIMPLEX, 2, box_color, 3, cv2.LINE_AA)

# 테스트 코드
# xy_list = [[321.0, 213.0, 401.0, 462.0], [432.0, 239.0, 495.0, 442.0], [472.0, 199.0, 525.0, 402.0]]

# # 가장 가까운 두 박스 찾기
# closest_boxes = find_closest_pairs(xy_list)
# print("가장 가까운 두 박스:", closest_boxes)

# # 두 박스를 포함하는 가장 큰 박스 생성
# largest_box = create_largest_box(*closest_boxes)
# print("가장 큰 박스의 좌표:", largest_box)
```

```python
import cv2
from ultralytics import YOLO
import numpy as np
from tensorflow.keras.models import load_model
from PIL import ImageFont, ImageDraw, Image

# 영상 파일 위치 지정
video_file = 'mp4/fight06.mp4'
# yolo 모델 정의
# 학습 당시에는 v8x로 진행 하였음
best_model = 'yolov8n_pose.pt'
yolo_model = YOLO(best_model)
# lstm 모델 정의
lstm_model = load_model('lstm_se8.h5')

# # 시작 프레임 값 정의
# frame_idx = 0
# 시퀀스 정의
sequences = []

# 시퀀스 길이
sequence_length = 8

# 등장하는 동작 각각 0번 1번 2번 3번
actions = ["Fighting", "falldown", "standing", "walking"]

action_times = {action: None for action in actions}

dangerous_actions = ["Fighting", "falldown"]

# 이전에 그려진 박스의 좌표를 저장할 변수
prev_box = None

# 동영상 파일로부터 비디오 캡처 객체 생성
cap = cv2.VideoCapture(video_file)  
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

# 창 이름 설정
window_name = "Video"

# 창 생성 (사이즈 조절 가능하도록 WINDOW_NORMAL 사용)
cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

# 창의 크기 조정
cv2.resizeWindow(window_name, 640, 480)

while cap.isOpened():
    # 이미지의 프레임을 가져오는 코드
    success, frame = cap.read()
    # 영상 끝에 도달하면 끝나게 설정
    if not success:
        break  
    # 불러온 프레임에서 사람이 있으면 사람에 대한 바운딩 박스나 인체 키포인트 정보를 results에 저장 
    results = yolo_model(frame, save=False, classes=[0])
    # 바운딩 박스의 좌표를 리스트화 해주는 코드 (gpu로는 처리가 안되어 cpu로 보내어 처리)
    # xy_list = results[0].boxes.xyxy.cpu().numpy().tolist()

    # 오류 방지 코드
    if not results:
        continue

    # 인체 키포인트 정보 가져옴
    results_keypoint = results[0].keypoints.xyn.cpu().numpy()


    # 한명일때 한명에 대한 데이터를 저장하고 나머지 24개의 데이터를 0으로 채움 
    if len(results_keypoint) == 1:
        keypoint_list = extract_keypoint(results_keypoint[0])
        keypoint_list.extend([0] * 24)  
    # 두명 이상 일때 xy좌표를 읽어서 가장 가까운 두명에 대해서 좌표를 가져옴
    elif len(results_keypoint) >= 2:
        xy_list = results[0].boxes.xyxy.cpu().numpy().tolist()
        closest_indices = find_closest_pair_indices(xy_list)
        
        keypoint_list1 = extract_keypoint(results_keypoint[0])
        keypoint_list2 = extract_keypoint(results_keypoint[1])
        keypoint_list = keypoint_list1 + keypoint_list2

    # 오류 방지 코드
    else:
        continue     

    # 시퀀스 리스트에 위에서 가져온 데이터들을 저장함
    sequences.append(keypoint_list)
    # 20쌍의 데이터가 모이면 lstm 모델에 통과 시켜 어떤 동장을 수행하는지 여부를 추출함
    if len(sequences) == sequence_length:
        sequences_array = np.array(sequences).reshape(1, sequence_length, -1)
        predictions = lstm_model.predict(sequences_array)
        # print(predictions)

        # 예측 값중 가장 큰 값을 찾기
        action_index = np.argmax(predictions)
        action_name = actions[action_index]
        # 가장 높은 가능성의 퍼센트 출력
        max_value = np.max(predictions)


        if action_name in dangerous_actions:
            last_dangerous_action_detected = action_name
        else:
            last_dangerous_action_detected = None

        # Display predicted action
        if last_dangerous_action_detected and max_value > 0.9:
            box_color = (0, 0, 255)  # Red color for dangerous actions
            # if last_dangerous_action_detected == "falldown":
            #     cv2.putText(frame, 'Danger! Someone has fallen!', (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3, cv2.LINE_AA)
            # else:
            #     cv2.putText(frame, 'Danger! Assault detected!', (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1 , (0, 0, 255), 3, cv2.LINE_AA)
        elif last_dangerous_action_detected and max_value > 0.5:
            box_color = (0, 128, 255)  # Red color for dangerous actions
        else:
            box_color = (0, 255, 0)  # Green color for non-dangerous actions

        # 바운딩 박스 좌표 구하기
        xy_list = results[0].boxes.xyxy.cpu().numpy().tolist()
        closest_boxes = find_closest_pairs(xy_list)
        largest_box = create_largest_box(*closest_boxes)
        # cv2.rectangle(frame,(int(largest_box[0]), int(largest_box[1])), (int(largest_box[2]), int(largest_box[3])), box_color, 2)
        # cv2.putText(frame, action_name, (int(calculate_font_center(largest_box)[0]), int(calculate_font_center(largest_box)[1])), cv2.FONT_HERSHEY_SIMPLEX, 2, box_color, 3, cv2.LINE_AA)
        prev_box = largest_box
        sequences.clear()

    else:
        if prev_box is not None:  # 이전에 그려진 박스가 있으면
            cv2.rectangle(frame,(int(prev_box[0]), int(prev_box[1])), (int(prev_box[2]), int(prev_box[3])), box_color, 2)
            cv2.putText(frame, action_name, (int(prev_box[0]), int(prev_box[3]) + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, box_color, 2, cv2.LINE_AA)
            if last_dangerous_action_detected and max_value > 0.9:
                if last_dangerous_action_detected == "falldown":
                    cv2.putText(frame, 'Danger! Someone has fallen!', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3, cv2.LINE_AA)
                else:
                    cv2.putText(frame, 'Danger! Assault detected!', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3, cv2.LINE_AA)



    cv2.imshow(window_name, frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Exit when ‘q’ is pressed
        break




cap.release()
cv2.destroyAllWindows()
```
